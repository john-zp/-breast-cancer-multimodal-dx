```python
import pandas as pd
```

```python
bfeature1 = pd.read_excel('bfeature1.xlsx')
bfeature1
```

```python
bfeature2 = pd.read_excel('bfeature2.xlsx')
bfeature2
```

```python
bfeature = pd.concat([bfeature1, bfeature2])
bfeature
```

```python
bfeature.set_index('ID',inplace=True)
bfeature
```

```python
bfeature.to_excel('03_bfeature.xlsx')
```

```python
train_bfeature = bfeature.sample(frac=0.8, random_state=99)
train_bfeature
```

```python
train_bfeature.to_excel('03_train_bfeature.xlsx')
```

```python
train_bfeature.index
```

```python
~bfeature.index.isin(train_bfeature.index)
```

```python
test_bfeature = bfeature[~bfeature.index.isin(train_bfeature.index)]
test_bfeature
```

```python
test_bfeature.to_excel('03_test_bfeature.xlsx')
```
