# LASSO

```python
param_grid={'C': np.logspace(-3,3,1000)}
```

```python
log_l1 = LogisticRegression(solver='liblinear',random_state=99,penalty='l1')
```

```python
grid_search = GridSearchCV(log_l1, param_grid)
```

```python
grid_search.fit(train_qfeature_s, train_target)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
grid_search.best_estimator_.score(test_qfeature_s, test_target)
```

```python
lasso_qfeature = grid_search.best_estimator_.coef_!=0
lasso_qfeature = lasso_qfeature.reshape(lasso_qfeature.shape[1])
```

```python
lasso_qfeature
```

```python
train_qfeature_.columns[lasso_qfeature]
```

```python
train_qfeature__ = train_qfeature_.columns[lasso_qfeature]
train_qfeature__ = train_qfeature_[train_qfeature__]
```

```python
test_qfeature__ = test_qfeature_.columns[lasso_qfeature]
test_qfeature__ = test_qfeature_[test_qfeature__]
```

```python
train_qfeature__s = sstransfer.fit_transform(train_qfeature__)
test_qfeature__s = sstransfer.transform(test_qfeature__)
```

```python
log_ = LogisticRegression(solver='newton-cg',random_state=99, penalty= None)
```

```python
log_.fit(train_qfeature__s, train_target)
```

```python
log_.coef_
```

```python
log_.score(train_qfeature__s,train_target)
```

```python
log_.score(test_qfeature__s,test_target)
```

```python
from sklearn.utils import resample
```

```python

n_iterations = 1000
train_scores = []

for _ in range(n_iterations):
    X_sample, y_sample = resample(train_qfeature__s, train_target)
    score = log_.score(X_sample, y_sample)
    train_scores.append(score)

train_lower = np.percentile(train_scores, 2.5)
train_upper = np.percentile(train_scores, 97.5)

print(f"The 95% confidence interval of the accuracy rate of the training set: ({train_lower:.3f}, {train_upper:.3f})")


test_scores = []

for _ in range(n_iterations):
    X_sample, y_sample = resample(test_qfeature__s, test_target)
    score = log_.score(X_sample, y_sample)
    test_scores.append(score)

test_lower = np.percentile(test_scores, 2.5)
test_upper = np.percentile(test_scores, 97.5)

print(f"The 95% confidence interval of the accuracy rate of the testing set: ({test_lower:.3f}, {test_upper:.3f})")
```

```python
from sklearn.metrics import roc_auc_score
```

```python

y_train_prob = log_.predict_proba(train_qfeature__s)[:, 1]


auc_train = roc_auc_score(train_target, y_train_prob)

print(f"TARIN AUC: {auc_train:.3f}")
```

```python

y_test_prob = log_.predict_proba(test_qfeature__s)[:, 1]


auc_test = roc_auc_score(test_target, y_test_prob)

print(f"TEST AUC: {auc_test:.3f}")
```

```python
n_iterations = 1000
train_auc_scores = []
test_auc_scores = []


for _ in range(n_iterations):
    X_sample, y_sample = resample(train_qfeature__s, train_target)
    y_pred_proba = log_.predict_proba(X_sample)[:, 1]
    auc = roc_auc_score(y_sample, y_pred_proba)
    train_auc_scores.append(auc)

train_lower = np.percentile(train_auc_scores, 2.5)
train_upper = np.percentile(train_auc_scores, 97.5)
print(f"TRAIN AUC çš„ 95% CI: ({train_lower:.3f}, {train_upper:.3f})")

for _ in range(n_iterations):
    X_sample, y_sample = resample(test_qfeature__s, test_target)
    y_pred_proba = log_.predict_proba(X_sample)[:, 1]
    auc = roc_auc_score(y_sample, y_pred_proba)
    test_auc_scores.append(auc)

test_lower = np.percentile(test_auc_scores, 2.5)
test_upper = np.percentile(test_auc_scores, 97.5)
print(f"TEST AUC çš„ 95% CI: ({test_lower:.3f}, {test_upper:.3f})")
```
