# Train the model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import plot_confusion_matrix, plot_roc_curve, plot_precision_recall_curve
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV,cross_val_score
```

# LG

```python
log = LogisticRegression(solver='newton-cg',random_state=99, penalty='none')
```

```python
log.fit(train_bfeature_, train_target)
```

```python
log.coef_
```

```python
log.intercept_
```

```python
log.score(train_bfeature_,train_target)
```

```python
log.score(test_bfeature_,test_target)
```

```python
plot_roc_curve(log, test_bfeature_, test_target)
```

```python
plot_precision_recall_curve(log, test_bfeature_, test_target)
```

```python
param_grid={'C': np.logspace(-3,3,1000)}
```

```python
log_l2 = LogisticRegression(random_state=99,penalty='l2')
```

```python
grid_search = GridSearchCV(log_l2, param_grid)
```

```python
grid_search.fit(train_bfeature_,train_target)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
grid_search.best_estimator_.score(test_bfeature_,test_target)
```

```python
log_l1 = LogisticRegression(solver='liblinear',random_state=99,penalty='l1')
```

```python
grid_search_ = GridSearchCV(log_l1, param_grid)
```

```python
grid_search_.fit(train_bfeature_,train_target)
```

```python
grid_search_.best_params_
```

```python
grid_search_.best_score_
```

```python
grid_search_.best_estimator_.score(test_bfeature_,test_target)
```
