# Sort the file names and bind them with the ID numbers (all files, including non-standard materials)

```python
import os
```

```python
filePath = 'E:/CSTKZ/ORI/bi_rads_internal_images'
```

```python
i_=''
j_=[]
k_=[]
for i, j, k in os.walk(filePath):
    j_.append(j)
    k_.append(k)
```

```python
import numpy as np
```

```python
j_ = np.array(j_)
j_
```

```python
id_ = j_[0]
id_
```

```python
k_ = np.array(k_)
k_
```

```python
name_ = k_[1:]
name_
```

```python
dict_ = dict(zip(id_, name_))
dict_
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(dict_.values())
df
```

```python
df.index = dict_.keys()
df
```

```python
df.to_excel("01_1_part1.xlsx")
```
