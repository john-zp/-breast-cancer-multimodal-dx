# RF

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV,cross_val_score
```

# Calibration curve

```python

def proportion_ci(p, n, z=1.96):
    if n == 0:
        return np.nan, np.nan
    se = np.sqrt(p * (1 - p) / n)
    return round(max(0, p - z * se), 3), round(min(1, p + z * se), 3)


bi_rads = ['BI-RADS 2', 'BI-RADS 3', 'BI-RADS 4A', 'BI-RADS 4B', 'BI-RADS 4C', 'BI-RADS 5']
re = pd.DataFrame()
train_score = []
test_score = []
train_counts = []
test_counts = []

for bi in bi_rads:
    
    train_mask = train_bqrfeature[bi] == 1
    test_mask = test_bqrfeature[bi] == 1

    train_bqrfeature_bi = train_bqrfeature_[train_mask]
    train_target_bi = train_target[train_mask]
    test_bqrfeature_bi = test_bqrfeature_[test_mask]
    test_target_bi = test_target[test_mask]

    train_bi_total = train_mask.sum()
    test_bi_total = test_mask.sum()
    train_counts.append(train_bi_total)
    test_counts.append(test_bi_total)

    
    if bi in ['BI-RADS 2', 'BI-RADS 3', 'BI-RADS 4A', 'BI-RADS 5']:
        rfc_clf_bi = RandomForestClassifier(n_estimators=500, random_state=99,
                                            min_samples_split=4, min_samples_leaf=16,
                                            class_weight='balanced')
    else:
        rfc_clf_bi = RandomForestClassifier(n_estimators=500, random_state=99,
                                            min_samples_split=4, min_samples_leaf=16)
    
    rfc_clf_bi.fit(train_bqrfeature_bi, train_target_bi)

    
    train_pred = rfc_clf_bi.predict(train_bqrfeature_bi)
    test_pred = rfc_clf_bi.predict(test_bqrfeature_bi)
    train_sc = accuracy_score(train_target_bi, train_pred)
    test_sc = accuracy_score(test_target_bi, test_pred)
    train_score.append(train_sc)
    test_score.append(test_sc)

    
    train_true_proba = train_target_bi.sum() / train_bi_total
    train_pred_proba = train_pred.sum() / train_bi_total
    test_true_proba = test_target_bi.sum() / test_bi_total
    test_pred_proba = test_pred.sum() / test_bi_total

    re_ = pd.DataFrame([[train_true_proba, train_pred_proba, test_true_proba, test_pred_proba]])
    re = pd.concat((re, re_), ignore_index=True)


train_ci = [proportion_ci(p, n) for p, n in zip(train_score, train_counts)]
test_ci = [proportion_ci(p, n) for p, n in zip(test_score, test_counts)]


score = pd.DataFrame({
    "BI-RADS": bi_rads,
    "Train Accuracy": train_score,
    "Train CI Lower": [ci[0] for ci in train_ci],
    "Train CI Upper": [ci[1] for ci in train_ci],
    "Test Accuracy": test_score,
    "Test CI Lower": [ci[0] for ci in test_ci],
    "Test CI Upper": [ci[1] for ci in test_ci],
})


re.columns = ['Train True Proba', 'Train Pred Proba', 'Test True Proba', 'Test Pred Proba']
re['BI-RADS'] = bi_rads
```

```python
bi_rads = ['BI-RADS 2', 'BI-RADS 3', 'BI-RADS 4A', 'BI-RADS 4B', 'BI-RADS 4C' ,'BI-RADS 5']
re = pd.DataFrame()
train_score = []
test_score = []

for bi in bi_rads:
  
    train_bqrfeature_bi = train_bqrfeature_[train_bqrfeature[bi]==1]
    train_target_bi = train_target[train_bqrfeature[bi]==1]
   
    test_bqrfeature_bi = test_bqrfeature_[test_bqrfeature[bi]==1]
    test_target_bi = test_target[test_bqrfeature[bi]==1]

    
    
   
    if bi in ['BI-RADS 2', 'BI-RADS 3', 'BI-RADS 4A', 'BI-RADS 5']:
        rfc_clf_bi = RandomForestClassifier(n_estimators=500,random_state=99,min_samples_split=4,min_samples_leaf=16,class_weight='balanced')
        rfc_clf_bi.fit(train_bqrfeature_bi, train_target_bi)
        train_bi_score = rfc_clf_bi.score(train_bqrfeature_bi, train_target_bi)
        test_bi_score = rfc_clf_bi.score(test_bqrfeature_bi, test_target_bi)
        train_score.append(train_bi_score)
        test_score.append(test_bi_score)
    else:
        rfc_clf_bi = RandomForestClassifier(n_estimators=500,random_state=99,min_samples_split=4,min_samples_leaf=16)
        rfc_clf_bi.fit(train_bqrfeature_bi, train_target_bi)
        train_bi_score = rfc_clf_bi.score(train_bqrfeature_bi, train_target_bi)
        test_bi_score = rfc_clf_bi.score(test_bqrfeature_bi, test_target_bi)
        train_score.append(train_bi_score)
        test_score.append(test_bi_score)
    
   
    train_bi_total = train_bqrfeature[bi].sum()
    train_true_sum = train_target[train_bqrfeature[bi]==1].sum()
    train_true_proba = train_true_sum / train_bi_total
   
    train_pre_sum = rfc_clf_bi.predict(train_bqrfeature_bi).sum()
    train_pre_proba = train_pre_sum / train_bi_total
    
    
    test_bi_total = test_bqrfeature[bi].sum()
    test_true_sum = test_target[test_bqrfeature[bi]==1].sum()
    test_true_proba = test_true_sum / test_bi_total
      
    test_pre_sum = rfc_clf_bi.predict(test_bqrfeature_bi).sum()
    test_pre_proba = test_pre_sum / test_bi_total
    
    re_ = pd.DataFrame((train_true_proba, train_pre_proba, test_true_proba, test_pre_proba))
    re = pd.concat((re, re_.T))
```
