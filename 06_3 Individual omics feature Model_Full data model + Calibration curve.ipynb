## T-TEST

```python
train_rfeature_B1 = train_rfeature.drop(columns=['IMG','BI-RADS 2', 'BI-RADS 3', 'BI-RADS 4A', 'BI-RADS 4B', 'BI-RADS 4C' ,'BI-RADS 5'])[train_rfeature['B']==1]
train_rfeature_B1
```

```python
train_rfeature_B0 = train_rfeature.drop(columns=['IMG','BI-RADS 2', 'BI-RADS 3', 'BI-RADS 4A', 'BI-RADS 4B', 'BI-RADS 4C' ,'BI-RADS 5'])[train_rfeature['B']==0]
train_rfeature_B0
```

```python
columns_ = train_rfeature_.columns
from scipy.stats import ttest_ind, levene
features = []
for col in columns_:
    if levene(train_rfeature_B0[col], train_rfeature_B1[col])[1] > 0.05:
        if ttest_ind(train_rfeature_B0[col], train_rfeature_B1[col])[1] < 0.05:
            features.append(col)
    else:
        if ttest_ind(train_rfeature_B0[col], train_rfeature_B1[col], equal_var=False)[1] < 0.05:
             features.append(col)
```

```python
train_rfeature__ = train_rfeature_[features]
train_rfeature__
```

```python
test_rfeature__ = test_rfeature_[features]
test_rfeature__
```

# Calibration curve

```python
bi_rads = ['BI-RADS 2', 'BI-RADS 3', 'BI-RADS 4A', 'BI-RADS 4B', 'BI-RADS 4C' ,'BI-RADS 5']
```

```python
score = pd.DataFrame()
re = pd.DataFrame()
for bi in bi_rads:
    train_bi_total = train_rfeature[bi].sum()
    train_true_sum = train_target[train_rfeature[bi]==1].sum()
    train_true_proba = train_true_sum / train_bi_total
    train_pre_sum = rfc_clf.predict(train_rfeature__sd[train_rfeature[bi]==1]).sum()
    train_pre_proba = train_pre_sum / train_bi_total
    train_sc = accuracy_score(train_target[train_rfeature[bi]==1],rfc_clf.predict(train_rfeature__sd[train_rfeature[bi]==1]))
    
    
    test_bi_total = test_rfeature[bi].sum()
    test_true_sum = test_target[test_rfeature[bi]==1].sum()
    test_true_proba = test_true_sum / test_bi_total
       
    test_pre_sum = rfc_clf.predict(test_rfeature__sd[test_rfeature[bi]==1]).sum()
    test_pre_proba = test_pre_sum / test_bi_total
    
    test_sc = accuracy_score(test_target[test_rfeature[bi]==1],rfc_clf.predict(test_rfeature__sd[test_rfeature[bi]==1]))
        
    score_ = pd.DataFrame((train_sc, test_sc))
    score = pd.concat((score, score_.T))

    
    re_ = pd.DataFrame((train_true_proba, train_pre_proba, test_true_proba, test_pre_proba))
    re = pd.concat((re, re_.T))
```

```python

def proportion_ci(p, n, z=1.96):
    if n == 0:
        return np.nan, np.nan
    se = np.sqrt(p * (1 - p) / n)
    lower = max(0, p - z * se)
    upper = min(1, p + z * se)
    return round(lower, 3), round(upper, 3)


score = pd.DataFrame()
re = pd.DataFrame()
train_counts = []
test_counts = []
train_scores = []
test_scores = []

for bi in bi_rads:
   
    train_mask = train_rfeature[bi] == 1
    test_mask = test_rfeature[bi] == 1

    
    train_bi_total = train_mask.sum()
    test_bi_total = test_mask.sum()
    train_counts.append(train_bi_total)
    test_counts.append(test_bi_total)

    
    train_y_true = train_target[train_mask]
    train_y_pred = rfc_clf.predict(train_rfeature__sd[train_mask])
    train_sc = accuracy_score(train_y_true, train_y_pred)
    train_true_proba = train_y_true.sum() / train_bi_total
    train_pre_proba = train_y_pred.sum() / train_bi_total
    train_scores.append(train_sc)

   
    test_y_true = test_target[test_mask]
    test_y_pred = rfc_clf.predict(test_rfeature__sd[test_mask])
    test_sc = accuracy_score(test_y_true, test_y_pred)
    test_true_proba = test_y_true.sum() / test_bi_total
    test_pre_proba = test_y_pred.sum() / test_bi_total
    test_scores.append(test_sc)

    
    re_ = pd.DataFrame([[train_true_proba, train_pre_proba, test_true_proba, test_pre_proba]])
    re = pd.concat((re, re_), ignore_index=True)


train_ci = [proportion_ci(p, n) for p, n in zip(train_scores, train_counts)]
test_ci = [proportion_ci(p, n) for p, n in zip(test_scores, test_counts)]


score = pd.DataFrame({
    "BI-RADS": bi_rads,
    "Train Accuracy": train_scores,
    "Train CI Lower": [ci[0] for ci in train_ci],
    "Train CI Upper": [ci[1] for ci in train_ci],
    "Test Accuracy": test_scores,
    "Test CI Lower": [ci[0] for ci in test_ci],
    "Test CI Upper": [ci[1] for ci in test_ci],
})
```
