# Unsupervised classification

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
mmtransfer = MinMaxScaler()
```

```python
mmfeatures = mmtransfer.fit_transform(brfeature_Mr)
```

```python
sstransfer = StandardScaler()
```

```python
ssfeatures = sstransfer.fit_transform(brfeature_Mr)
```

```python
from sklearn.decomposition import PCA
```

```python
pcatransfer = PCA(n_components=2)
```

```python
pcafeature = pcatransfer.fit_transform(ssfeatures)
pcafeature
```

```python
x=pcafeature[:,0]
y=pcafeature[:,1]
```

```python
plt.scatter(x,y)
```

```python
from sklearn.cluster import KMeans
```

```python
clusters = np.arange(2,11)
clusters
```

```python
inertia=[]
for i in clusters:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(mmfeatures)
    inertia.append(kmeans.inertia_)
```

```python
inertia
```

```python
plt.plot(clusters,inertia,"bo-")
```
