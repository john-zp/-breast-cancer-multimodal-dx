## T-TEST

```python
train_rfeature_B1 = train_rfeature.drop(columns=['IMG'])[train_rfeature['B']==1]
train_rfeature_B1
```

```python
train_rfeature_B0 = train_rfeature.drop(columns=['IMG'])[train_rfeature['B']==0]
train_rfeature_B0
```

```python
columns_ = train_rfeature_.columns
from scipy.stats import ttest_ind, levene
features = []
for col in columns_:
    if levene(train_rfeature_B0[col], train_rfeature_B1[col])[1] > 0.05:
        if ttest_ind(train_rfeature_B0[col], train_rfeature_B1[col])[1] < 0.05:
            features.append(col)
    else:
        if ttest_ind(train_rfeature_B0[col], train_rfeature_B1[col], equal_var=False)[1] < 0.05:
             features.append(col)
```

```python
train_rfeature__ = train_rfeature_[features]
train_rfeature__
```

```python
test_rfeature__ = test_rfeature_[features]
test_rfeature__
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.utils import resample
```

```python

y_train_prob = log_l2.predict_proba(train_rfeature__sd)[:, 1]


auc_train = roc_auc_score(train_target, y_train_prob)

print(f"TRAIN AUC: {auc_train:.3f}")


y_test_prob = log_l2.predict_proba(test_rfeature__sd,)[:, 1]


auc_test = roc_auc_score(test_target, y_test_prob)

print(f"TEST AUC: {auc_test:.3f}")
```

```python

n_iterations = 1000
train_scores = []

for _ in range(n_iterations):
    X_sample, y_sample = resample(train_rfeature__sd, train_target)
    score = log_l2.score(X_sample, y_sample)
    train_scores.append(score)

train_lower = np.percentile(train_scores, 2.5)
train_upper = np.percentile(train_scores, 97.5)

print(f"The 95% confidence interval of the accuracy rate of the training set: ({train_lower:.3f}, {train_upper:.3f})")


test_scores = []

for _ in range(n_iterations):
    X_sample, y_sample = resample(test_rfeature__sd, test_target)
    score = log_l2.score(X_sample, y_sample)
    test_scores.append(score)

test_lower = np.percentile(test_scores, 2.5)
test_upper = np.percentile(test_scores, 97.5)

print(f"The 95% confidence interval of the accuracy rate of the testing set: ({test_lower:.3f}, {test_upper:.3f})")
```
