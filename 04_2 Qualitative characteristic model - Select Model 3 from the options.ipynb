# IMPORT DATA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_bfeature = pd.read_excel('03_train_bfeature.xlsx',index_col=0)
test_bfeature = pd.read_excel('03_test_bfeature.xlsx', index_col=0)
```

# Data set division

```python
train_bfeature_ = train_bfeature.iloc[:,5:39]
train_bfeature_.head()
```

```python
test_bfeature_ = test_bfeature.iloc[:,5:39]
test_bfeature_.head()
```

```python
train_target = train_bfeature.B
train_target.head()
```

```python
test_target = test_bfeature.B
test_target.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import GridSearchCV,RandomizedSearchCV,cross_val_score
```

```python
log = LogisticRegression(solver='newton-cg',random_state=99, penalty=None)
```

```python
log.fit(train_bfeature_, train_target)
```

```python
log.coef_
```

```python
log.intercept_
```

```python
log.score(train_bfeature_,train_target)
```

```python
log.score(test_bfeature_,test_target)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
from sklearn.utils import resample
```

```python
y_train_prob = log.predict_proba(train_bfeature_)[:, 1]

auc_train = roc_auc_score(train_target, y_train_prob)

print(f"TRAIN AUC: {auc_train:.3f}")

y_test_prob = log.predict_proba(test_bfeature_,)[:, 1]

auc_test = roc_auc_score(test_target, y_test_prob)

print(f"TEST AUC: {auc_test:.3f}")
```

```python

n_iterations = 1000
train_scores = []

for _ in range(n_iterations):
    X_sample, y_sample = resample(train_bfeature_, train_target)
    score = log.score(X_sample, y_sample)
    train_scores.append(score)

train_lower = np.percentile(train_scores, 2.5)
train_upper = np.percentile(train_scores, 97.5)

print(f"The 95% confidence interval of the accuracy rate of the training set: ({train_lower:.3f}, {train_upper:.3f})")


test_scores = []

for _ in range(n_iterations):
    X_sample, y_sample = resample(test_bfeature_, test_target)
    score = log.score(X_sample, y_sample)
    test_scores.append(score)

test_lower = np.percentile(test_scores, 2.5)
test_upper = np.percentile(test_scores, 97.5)

print(f"The 95% confidence interval of the accuracy rate of the testing set: ({test_lower:.3f}, {test_upper:.3f})")
```

```python
n_iterations = 1000
train_auc_scores = []
test_auc_scores = []

for _ in range(n_iterations):
    X_sample, y_sample = resample(train_bfeature_, train_target)
    y_pred_proba = log.predict_proba(X_sample)[:, 1]
    auc = roc_auc_score(y_sample, y_pred_proba)
    train_auc_scores.append(auc)

train_lower = np.percentile(train_auc_scores, 2.5)
train_upper = np.percentile(train_auc_scores, 97.5)
print(f"The 95% confidence interval of the AUC of the training set: ({train_lower:.3f}, {train_upper:.3f})")


for _ in range(n_iterations):
    X_sample, y_sample = resample(test_bfeature_, test_target)
    y_pred_proba = log.predict_proba(X_sample)[:, 1]
    auc = roc_auc_score(y_sample, y_pred_proba)
    test_auc_scores.append(auc)

test_lower = np.percentile(test_auc_scores, 2.5)
test_upper = np.percentile(test_auc_scores, 97.5)
print(f"The 95% confidence interval of the AUC of the testing set: ({test_lower:.3f}, {test_upper:.3f})")
```
