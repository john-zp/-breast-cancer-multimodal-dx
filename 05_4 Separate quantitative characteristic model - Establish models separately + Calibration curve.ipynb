```python
bi_rads = ['BI-RADS 3', 'BI-RADS 4A', 'BI-RADS 4B', 'BI-RADS 4C' ,'BI-RADS 5']
re = pd.DataFrame()
train_score = []
test_score = []
train_auc= []
test_auc= []

for bi in bi_rads:
    train_qfeature_bi = train_qfeature__s[train_bqfeature[bi]==1]
    train_target_bi = train_target[train_bqfeature[bi]==1]
    test_qfeature_bi = test_qfeature__s[test_bqfeature[bi]==1]
    test_target_bi = test_target[test_bqfeature[bi]==1]

    
    
    if bi in ['BI-RADS 2', 'BI-RADS 3', 'BI-RADS 4A', 'BI-RADS 5']:
        rfc_clf_bi = RandomForestClassifier(n_estimators=500,random_state=99,min_samples_split=16,min_samples_leaf=7, class_weight='balanced')
        rfc_clf_bi.fit(train_qfeature_bi, train_target_bi)
        train_bi_score = rfc_clf_bi.score(train_qfeature_bi, train_target_bi)
        test_bi_score = rfc_clf_bi.score(test_qfeature_bi, test_target_bi)
    else:
        rfc_clf_bi = RandomForestClassifier(n_estimators=500,random_state=99,min_samples_split=16,min_samples_leaf=7)
        rfc_clf_bi.fit(train_qfeature_bi, train_target_bi)
        train_bi_score = rfc_clf_bi.score(train_qfeature_bi, train_target_bi)
        test_bi_score = rfc_clf_bi.score(test_qfeature_bi, test_target_bi)
 
    
    train_bi_auc = roc_auc_score(train_target_bi,rfc_clf_bi.predict_proba(train_qfeature_bi)[:,1])
    test_bi_auc = roc_auc_score(test_target_bi,rfc_clf_bi.predict_proba(test_qfeature_bi)[:,1])
    
    train_score.append(train_bi_score)
    test_score.append(test_bi_score)
    
    train_auc.append(train_bi_auc)
    test_auc.append(test_bi_auc)

    
    
    train_bi_total = train_bqfeature[bi].sum()
    train_true_sum = train_target[train_bqfeature[bi]==1].sum()
    train_true_proba = train_true_sum / train_bi_total
   
    train_pre_sum = rfc_clf_bi.predict(train_qfeature_bi).sum()
    train_pre_proba = train_pre_sum / train_bi_total
    
    
    test_bi_total = test_bqfeature[bi].sum()
    test_true_sum = test_target[test_bqfeature[bi]==1].sum()
    test_true_proba = test_true_sum / test_bi_total    
    test_pre_sum = rfc_clf_bi.predict(test_qfeature_bi).sum()
    test_pre_proba = test_pre_sum / test_bi_total
    
    re_ = pd.DataFrame((train_true_proba, train_pre_proba, test_true_proba, test_pre_proba))
    re = pd.concat((re, re_.T))
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier

def proportion_ci(p, n, z=1.96):
    if n == 0:
        return np.nan, np.nan
    se = np.sqrt(p * (1 - p) / n)
    lower = max(0, p - z * se)
    upper = min(1, p + z * se)
    return round(lower, 3), round(upper, 3)

bi_rads = ['BI-RADS 2', 'BI-RADS 3', 'BI-RADS 4A', 'BI-RADS 4B', 'BI-RADS 4C' ,'BI-RADS 5']
re = pd.DataFrame()
train_score = []
test_score = []
train_counts = []
test_counts = []

for bi in bi_rads:
    train_mask = train_bqfeature[bi] == 1
    test_mask = test_bqfeature[bi] == 1

    train_qfeature_bi = train_qfeature__s[train_mask]
    train_target_bi = train_target[train_mask]
    test_qfeature_bi = test_qfeature__s[test_mask]
    test_target_bi = test_target[test_mask]

    train_bi_total = train_mask.sum()
    test_bi_total = test_mask.sum()
    train_counts.append(train_bi_total)
    test_counts.append(test_bi_total)

    if bi in ['BI-RADS 2', 'BI-RADS 3', 'BI-RADS 4A', 'BI-RADS 5']:
        rfc_clf_bi = RandomForestClassifier(n_estimators=500, random_state=99,
                                            min_samples_split=16, min_samples_leaf=7,
                                            class_weight='balanced')
    else:
        rfc_clf_bi = RandomForestClassifier(n_estimators=500, random_state=99,
                                            min_samples_split=16, min_samples_leaf=7)

    rfc_clf_bi.fit(train_qfeature_bi, train_target_bi)

    train_bi_score = rfc_clf_bi.score(train_qfeature_bi, train_target_bi)
    test_bi_score = rfc_clf_bi.score(test_qfeature_bi, test_target_bi)
    train_score.append(train_bi_score)
    test_score.append(test_bi_score)

    train_true_sum = train_target_bi.sum()
    train_true_proba = train_true_sum / train_bi_total
    train_pre_sum = rfc_clf_bi.predict(train_qfeature_bi).sum()
    train_pre_proba = train_pre_sum / train_bi_total

    test_true_sum = test_target_bi.sum()
    test_true_proba = test_true_sum / test_bi_total
    test_pre_sum = rfc_clf_bi.predict(test_qfeature_bi).sum()
    test_pre_proba = test_pre_sum / test_bi_total


    re_ = pd.DataFrame([[train_true_proba, train_pre_proba, test_true_proba, test_pre_proba]])
    re = pd.concat((re, re_), ignore_index=True)

train_ci = [proportion_ci(p, n) for p, n in zip(train_score, train_counts)]
test_ci = [proportion_ci(p, n) for p, n in zip(test_score, test_counts)]

result = pd.DataFrame({
    "BI-RADS": bi_rads,
    "Train Accuracy": train_score,
    "Train CI Lower": [ci[0] for ci in train_ci],
    "Train CI Upper": [ci[1] for ci in train_ci],
    "Test Accuracy": test_score,
    "Test CI Lower": [ci[0] for ci in test_ci],
    "Test CI Upper": [ci[1] for ci in test_ci],
})


```
