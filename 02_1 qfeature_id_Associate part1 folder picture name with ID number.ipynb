# The quantitative features of the image are bound to the ID number.

```python
import pandas as pd
```

```python
part1 = pd.read_excel('01_1_part1.xlsx',index_col=0)
part1
```

```python
part2 = pd.read_excel('01_2_part2.xlsx',index_col=0)
part2
```

```python
part1.index.is_unique
```

```python
part2.index.is_unique
```

```python
part1.duplicated().value_counts()
```

```python
part2.duplicated().value_counts()
```

```python
part1.drop_duplicates(inplace = True)
part1
```

```python
part2.drop_duplicates(inplace = True)
part2
```

```python
imgid = pd.concat([part1,part2])
imgid
```

```python
imgid_ = imgid.stack()
imgid_ = imgid_.reset_index()
imgid_
```

```python
imgid_.columns = ['ID','NO','IMG']
del imgid_['NO']
imgid_
```

```python
imgid_.index.is_unique
```

```python
imgid_.duplicated().value_counts()
```

```python
imgid_.drop_duplicates(subset=['IMG'], inplace = True)
imgid_
```

```python
imgid_.set_index('IMG', inplace=True)
imgid_
```

```python
imgid_.to_excel('02_1_imgid_.xlsx')
```

```python
qfeature = pd.read_csv('qfeature.csv', index_col=0)
qfeature
```

```python
idqfeature = pd.concat([qfeature,imgid_], axis=1, join='inner')
idqfeature
```

```python
idqfeature.reset_index(inplace=True)
idqfeature.set_index('ID',inplace=True)
idqfeature
```

```python
idqfeature.to_excel('02_1_idqfeature.xlsx')
```
