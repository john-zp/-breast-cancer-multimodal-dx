# IMPORT DATA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_bfeature = pd.read_excel('03_train_bfeature.xlsx',index_col=0)
train_bfeature
```

```python
test_bfeature = pd.read_excel('03_test_bfeature.xlsx', index_col=0)
test_bfeature
```

```python
qfeature = pd.read_excel('05_lasso_idqfeature.xlsx')
qfeature
```

```python
train_rfeature = pd.read_excel('06_train_sdr.xlsx', index_col=0)
train_rfeature
```

```python
test_rfeature = pd.read_excel('06_test_sdr.xlsx', index_col=0)
test_rfeature
```

```python
train_qrfeature = pd.merge(train_rfeature,qfeature,left_on=train_rfeature.IMG,right_on=qfeature['index'],how='inner')
train_qrfeature.set_index('ID',inplace=True)
```

```python
train_qrfeature.drop(columns=['key_0','IMG','index'],inplace=True)
train_qrfeature
```

```python
test_qrfeature = pd.merge(test_rfeature,qfeature,left_on=test_rfeature.IMG,right_on=qfeature['index'],how='inner')
test_qrfeature.set_index('ID',inplace=True)
```

```python
test_qrfeature.drop(columns=['key_0','IMG','index'],inplace=True)
test_qrfeature
```

```python
train_bqrfeature = pd.merge(train_qrfeature,train_bfeature,left_on=train_qrfeature.index,right_on=train_bfeature.index,how='inner')
train_bqrfeature
```

```python
test_bqrfeature = pd.merge(test_qrfeature,test_bfeature,left_on=test_qrfeature.index,right_on=test_bfeature.index,how='inner')
test_bqrfeature
```

```python
train_bqrfeature_ = train_bqrfeature.drop(
    columns=['key_0','ER','PR','HER','B','M','BI-RADS 1','BI-RADS 2', 'BI-RADS 3', 'BI-RADS 4A', 'BI-RADS 4B', 'BI-RADS 4C' ,'BI-RADS 5', 'BI-RADS 6'])
test_bqrfeature_ = test_bqrfeature.drop(
    columns=['key_0','ER','PR','HER','B','M','BI-RADS 1','BI-RADS 2', 'BI-RADS 3', 'BI-RADS 4A', 'BI-RADS 4B', 'BI-RADS 4C' ,'BI-RADS 5', 'BI-RADS 6'])
```

```python
train_bqrfeature_.info()
```

```python
train_target = train_bqrfeature.B
test_target = test_bqrfeature.B
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV,cross_val_score
from sklearn.metrics import roc_auc_score
from sklearn.utils import resample
```

```python
from sklearn.preprocessing import StandardScaler
sstransfer = StandardScaler()
```

```python
train_bqrfeature_.columns = train_bqrfeature_.columns.astype(str)
test_bqrfeature_.columns = test_bqrfeature_.columns.astype(str)
```

```python
train_bqrfeature_s = sstransfer.fit_transform(train_bqrfeature_)
test_bqrfeature_s = sstransfer.transform(test_bqrfeature_)
```

```python
train_bqrfeature_s
```

```python
grid_search = GridSearchCV(rfc_clf, param_grid)
```

```python
grid_search.fit(train_bqrfeature_,train_target)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
grid_search.score(train_bqrfeature_,train_target)
```

```python
grid_search.best_estimator_.score(test_bqrfeature_,test_target)
```

# LG

```python
log = LogisticRegression(solver='newton-cg',random_state=99,penalty=None)
```

```python
log.fit(train_bqrfeature_s,train_target)
```

```python
log.score(train_bqrfeature_s,train_target)
```

```python
log.score(test_bqrfeature_s,test_target)
```

```python
y_train_prob = log.predict_proba(train_bqrfeature_s)[:, 1]

auc_train = roc_auc_score(train_target, y_train_prob)

print(f"TRAIN AUC: {auc_train:.3f}")


y_test_prob = log.predict_proba(test_bqrfeature_s,)[:, 1]

auc_test = roc_auc_score(test_target, y_test_prob)

print(f"TEST AUC: {auc_test:.3f}")
```

```python

n_iterations = 1000
train_scores = []

for _ in range(n_iterations):
    X_sample, y_sample = resample(train_bqrfeature_s, train_target)
    score = log.score(X_sample, y_sample)
    train_scores.append(score)

train_lower = np.percentile(train_scores, 2.5)
train_upper = np.percentile(train_scores, 97.5)

print(f"95% CI of accuracy on the training set: ({train_lower:.3f}, {train_upper:.3f})")


test_scores = []

for _ in range(n_iterations):
    X_sample, y_sample = resample(test_bqrfeature_s, test_target)
    score = log.score(X_sample, y_sample)
    test_scores.append(score)

test_lower = np.percentile(test_scores, 2.5)
test_upper = np.percentile(test_scores, 97.5)

print(f"95% CI of accuracy on the testing set: ({test_lower:.3f}, {test_upper:.3f})")
```

```python
n_iterations = 1000
train_auc_scores = []
test_auc_scores = []

for _ in range(n_iterations):
    X_sample, y_sample = resample(train_bqrfeature_s, train_target)
    y_pred_proba = log.predict_proba(X_sample)[:, 1]
    auc = roc_auc_score(y_sample, y_pred_proba)
    train_auc_scores.append(auc)

train_lower = np.percentile(train_auc_scores, 2.5)
train_upper = np.percentile(train_auc_scores, 97.5)
print(f"95% CI of AUC on the training set: ({train_lower:.3f}, {train_upper:.3f})")

# 测试集 AUC 的 bootstrap
for _ in range(n_iterations):
    X_sample, y_sample = resample(test_bqrfeature_s, test_target)
    y_pred_proba = log.predict_proba(X_sample)[:, 1]
    auc = roc_auc_score(y_sample, y_pred_proba)
    test_auc_scores.append(auc)

test_lower = np.percentile(test_auc_scores, 2.5)
test_upper = np.percentile(test_auc_scores, 97.5)
print(f"95% CI of AUC on the testing set: ({test_lower:.3f}, {test_upper:.3f})")
```
