# Calibration curve

```python
bi_rads = ['BI-RADS 2', 'BI-RADS 3', 'BI-RADS 4A', 'BI-RADS 4B', 'BI-RADS 4C' ,'BI-RADS 5']
```

```python
def proportion_ci(p, n, z=1.96):
    if n == 0:
        return np.nan, np.nan
    se = np.sqrt(p * (1 - p) / n)
    return max(0, p - z * se), min(1, p + z * se)

score_list = []
re_list = []

for bi in bi_rads:
    
    train_mask = train_bqrfeature[bi] == 1
    test_mask = test_bqrfeature[bi] == 1

    
    train_X = train_bqrfeature_[train_mask]
    train_y = train_target[train_mask]
    train_n = train_mask.sum()
    train_pred = rfc_clf.predict(train_X)
    train_acc = accuracy_score(train_y, train_pred)
    train_ci_low, train_ci_up = proportion_ci(train_acc, train_n)

    
    train_true_proba = train_y.sum() / train_n
    train_pred_proba = train_pred.sum() / train_n

 
    test_X = test_bqrfeature_[test_mask]
    test_y = test_target[test_mask]
    test_n = test_mask.sum()
    test_pred = rfc_clf.predict(test_X)
    test_acc = accuracy_score(test_y, test_pred)
    test_ci_low, test_ci_up = proportion_ci(test_acc, test_n)

    test_true_proba = test_y.sum() / test_n
    test_pred_proba = test_pred.sum() / test_n

   
    score_list.append({
        "BI-RADS": bi,
        "Train Accuracy": round(train_acc, 3),
        "Train CI Lower": round(train_ci_low, 3),
        "Train CI Upper": round(train_ci_up, 3),
        "Test Accuracy": round(test_acc, 3),
        "Test CI Lower": round(test_ci_low, 3),
        "Test CI Upper": round(test_ci_up, 3),
    })

   
    re_list.append({
        "BI-RADS": bi,
        "Train True Proba": round(train_true_proba, 3),
        "Train Pred Proba": round(train_pred_proba, 3),
        "Test True Proba": round(test_true_proba, 3),
        "Test Pred Proba": round(test_pred_proba, 3),
    })


score = pd.DataFrame(score_list)
re = pd.DataFrame(re_list)

```
